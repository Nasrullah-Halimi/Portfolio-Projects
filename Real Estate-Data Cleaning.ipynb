```python
# Import libraries.
import pandas as pd
import numpy as np

# Read the data. 
df = pd.read_csv(r'C:\Users\nasru\OneDrive\Desktop\Python-PRJ\RealState\RS.csv')
```

```python
# Look at the data.
df.head()
```

```python
# Shows the missing data per column. 
df.isnull().sum()
```

```python
# Remove the "For rent:" from column Title.
df['Title'] = df['Title'].str.replace('For rent:', '')
df['Title']
```

```python
# Remove "€" and "Per month" from column Price.
df['Price'] = df['Price'].str.replace('€','')
df['Price'] = df['Price'].str.replace(',','')
df['Price'] = df['Price'].str.replace('per month','')

df.head()
```

```python
# Remove the measurement unit (m²) from columns Living Area and Volume. 
df['Volume'] = df['Volume'].str.replace('m²','')
df['Living Area'] = df['Living Area'].str.replace('m²','')

df.head()
```

```python
# Remove "rooms" from column Number of rooms. 
df['Number of rooms'] = df['Number of rooms'].str.replace('rooms','')
df.head()
```

```python
# Looking at the unique values in column Garden. 
df['Garden'].value_counts()
```

```python
# Extract only 'Present' and 'Not present' - Omit units of measurement, parentheses and additional info. 
df['Garden'] = df['Garden'].astype(str)
df['Garden'] = df['Garden'].str.extract('([a-zA-Z ]+)', expand=False).str.strip()
df['Garden'].value_counts()
```

```python
# Replace the missing data in column Volume with its mean value. 
# Calculate the average of Volume column. 
vol_avg = df['Volume'].astype(float).mean(axis=0).__round__()
print("Average of volume is:", vol_avg)
```

```python
# Replace NaN by mean value. 
df['Volume'].replace(np.nan, vol_avg, inplace=True)
```

```python
# Calculate the mean of column Number of bedroom. 
Avg_broom = df['Number of bedroom'].astype(float).mean(axis=0).__round__()
print("Average of bedroom is:", Avg_broom)
```

```python
# Replace NaN by mean value. 
df['Number of bedroom'].replace(np.nan, Avg_broom, inplace=True)
```

```python
# Calculate the Average of column Num of floors. 
avg_floors = df['Num of floors'].astype(float).mean(axis=0).__round__()
print("Average of Number of floors is:", avg_floors)
```

```python
# Replace the NaN with mean value. 
df['Num of floors'].replace(np.nan, avg_floors, inplace=True)
```

```python
# Check what is the most frequent value in column Storeroom. 
df['Storeroom'].value_counts(normalize=True).mul(100).round(1).astype(str)+' %'
```

```python
# Replace the missing data with the most frequent value. 
df['Storeroom'].replace(np.nan, "Not present", inplace=True)
```

```python
# Final check for missing data. 
df.isnull().sum()
```

```python
# Check for duplicates.
print(df.duplicated().value_counts())
```

```python
# Delete the duplicates.
df.drop_duplicates(inplace=True)
```

```python
# Change data types.
df['Price'] = df['Price'].astype(int)
df['Living Area'] = df['Living Area'].astype(int)
df['Volume'] = df['Volume'].astype(int)
df['Number of rooms'] = df['Number of rooms'].astype(float)
df['Year of construction'] = pd.to_datetime(df['Year of construction'], format='%Y', errors='coerce')

df.dtypes
```

```python
df.head(3)
```

```python
# Remove the decimal points from these columns. 
df['Number of bedroom'] = df['Number of bedroom'].astype(int).replace('\.0','', regex=True)
df['Num of floors'] = df['Num of floors'].astype(int).replace('\.0','', regex=True)
df['Number of rooms'] = df['Number of rooms'].astype(int).replace('\.0','', regex=True)
df.head()
```

```python
# Splitting column title into two columns of property Name and City.
df[['Name', 'City']] = df.Title.str.split(",", expand=True)
df.head()
```

```python
# Move the columns "Name", "City" to the second and third position within the dataframe. 
second_column = df.pop('Name')
third_column = df.pop('City')
df.insert(1,'Name', second_column)
df.insert(2,'City', third_column)

df.head()
```

```python
# Delete the columns not needed. 
df.drop(['Title', 'Name', 'Type of house 1','Type of construction', 'Num of room', 'Status', 'Availability'], axis=1, inplace=True)
```

```python
# Rename columns.
df.rename(columns={"Living Area":"Living_area","Number of rooms": "Room", "Number of bedroom": "Bedroom", "Num of floors": "Floor", "Type of house":"Type", "Year of construction":"Year_built"}, inplace=True)
df.head()
```

```python
# Sort the data by Price in decending order.  
df.sort_values(by=['Price'], inplace=True, ascending=False)
```

```python
# Save the dataset to a new file.
pd = df.to_csv('RS-Cleaned.csv', index=False)
```
